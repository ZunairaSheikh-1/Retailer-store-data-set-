```python
df['Item'] = df['Item'].fillna(df['Item'].mode()[0]) #null values ko fill karna hai 
```

```python
df.isnull().sum()
```

```python
df['Price Per Unit'] = df['Price Per Unit'].fillna(df['Price Per Unit'].mean()) #null values ko fill karna hai 
```

```python
df['Quantity'] = df['Quantity'].fillna(df['Quantity'].mean()) #null values ko fill karna hai 
```

```python
df.isnull().sum()
```

```python
df['Total Spent'] = df['Total Spent'].fillna(df['Total Spent'].mean()) #null values ko fill karna hai 
 
```

```python
df.isnull().sum()
```

```python
df['Discount Applied'] = df['Discount Applied'].fillna(df['Discount Applied'].mean()) #null values ko fill karna hai 
```

```python
df.isnull().sum()
```

```python
top_items = df.groupby('Item')['Quantity'].sum().sort_values(ascending=False).head(10)
print(top_items) #Top 10 sold items
```

```python
import matplotlib.pyplot as plt

# Step 1: Count each payment method
payment_counts = df['Payment Method'].value_counts()

# Step 2: Define custom colors
colors = ['green', 'red', 'yellow']

# Step 3: Create Pie Chart
plt.figure(figsize=(7,7))
plt.pie(payment_counts, labels=payment_counts.index, autopct='%1.1f%%',
        startangle=90, colors=colors, shadow=True)

plt.title('Payment Method Distribution', fontsize=14)
plt.show()

```

```python
# Step 1: Find top 10 categories by total quantity sold
top_categories = df.groupby('Category')['Quantity'].sum().sort_values(ascending=False).head(10).reset_index()

# Step 2: Plot the distribution using Seaborn
plt.figure(figsize=(10,6))
sns.barplot(x='Quantity', y='Category', data=top_categories, color='seagreen',edgecolor='black', legend=False)

# Step 3: Beautify the chart
plt.title('Top 10 Item Categories by Quantity Sold', fontsize=14)
plt.xlabel('Total Quantity Sold')
plt.ylabel('Item Category')
plt.tight_layout()
plt.show()
```

```python
# Step 1: Find the top 10 most sold products
top_products = df.groupby('Item')['Quantity'].sum().sort_values(ascending=False).head(10).reset_index()

# Step 2: Plot the distribution using Seaborn
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(10,6))
sns.barplot(
    x='Quantity',
    y='Item',
    data=top_products,
    color='seagreen',             # clean yellow tone
    edgecolor='black'         # makes bars stand out
)

# Step 3: Beautify chart
plt.title('Top 10 Most Selling Products', fontsize=16, fontweight='bold', color='black')
plt.xlabel('Total Quantity Sold', fontsize=12)
plt.ylabel('Product Name', fontsize=12)
plt.grid(axis='x', linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Step 1: Convert 'Transaction Date' to datetime format
df['Transaction Date'] = pd.to_datetime(df['Transaction Date'])

# Step 2: Extract month (as Year-Month)
df['Month'] = df['Transaction Date'].dt.to_period('M').astype(str)

# Step 3: Group by month and calculate total sales
monthly_sales = df.groupby('Month')['Total Spent'].sum().reset_index()

# Step 4: Sort by month
monthly_sales['Month'] = pd.to_datetime(monthly_sales['Month'])
monthly_sales = monthly_sales.sort_values('Month')

# Step 5: Plot line graph
plt.figure(figsize=(10,6))
sns.lineplot(
    x='Month',
    y='Total Spent',
    data=monthly_sales,
    marker='o',
    color='seagreen',  # yellow line color
    linewidth=2.5
)

# Step 6: Beautify chart
plt.title('Monthly Sales Trend (Line Graph)', fontsize=16, fontweight='bold')
plt.xlabel('Month', fontsize=12)
plt.ylabel('Total Sales (PKR)', fontsize=12)
plt.xticks(rotation=45)
plt.grid(True, linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()


```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Step 1: Calculate total revenue per category
category_revenue = df.groupby('Category')['Total Spent'].sum().reset_index()

# Step 2: Sort by highest revenue
category_revenue = category_revenue.sort_values(by='Total Spent', ascending=False)

# Step 3: Plot bar graph
plt.figure(figsize=(10,6))
sns.barplot(
    x='Total Spent',
    y='Category',
    data=category_revenue,
    color='seagreen'  # you can change color to 'gold', 'skyblue', 'orange'
)

# Step 4: Beautify chart
plt.title('Category-wise Revenue Distribution', fontsize=16, fontweight='bold')
plt.xlabel('Total Revenue (PKR)', fontsize=12)
plt.ylabel('Product Category', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Step 1: Group by Location and calculate total revenue
sales_by_location_type = df.groupby('Location')['Total Spent'].sum().reset_index()

# Step 2: Create a polished bar chart
plt.figure(figsize=(8,5))
sns.barplot(
    x='Location',
    y='Total Spent',
    data=sales_by_location_type,
    color='lightseagreen'
)

# Step 3: Beautify the chart
plt.title('In-Store vs Online Revenue', fontsize=16, fontweight='bold', pad=15)
plt.xlabel('Sales Channel (Location Type)', fontsize=12)
plt.ylabel('Total Revenue (PKR)', fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.tight_layout()

# Step 4: Display
plt.show()

```
