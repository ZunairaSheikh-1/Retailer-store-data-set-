```python
#Group by Payment Method and sum the total spent
payment_method_distribution = df.groupby('Payment Method')['Total Spent'].sum()

#Define colors
colors = ['yellow', 'green', 'red']

#Plot a pie chart
plt.figure(figsize=(8, 8))
plt.pie(
    payment_method_distribution,
    labels=payment_method_distribution.index,
    autopct='%1.1f%%',
    colors=colors,
    startangle=90,
    wedgeprops={'edgecolor': 'black', 'linewidth': 1.5}
)

#chart
plt.title('Payment Method Distribution', fontsize=16, fontweight='bold')
plt.tight_layout()
plt.show()
```

```python
#Convert 'Transaction Date' to datetime format (if not already)
df['Transaction Date'] = pd.to_datetime(df['Transaction Date'])

#Extract year and month
df['Month'] = df['Transaction Date'].dt.to_period('M')

#Group by month and calculate total revenue
monthly_sales = df.groupby('Month')['Total Spent'].sum().reset_index()

#Convert period to string for better display
monthly_sales['Month'] = monthly_sales['Month'].astype(str)

#Plot line graph
plt.figure(figsize=(10,6))
sns.lineplot(x='Month', y='Total Spent', data=monthly_sales, marker='o', color='seagreen')

#Chart
plt.title('Monthly Sales Trends', fontsize=16, fontweight='bold')
plt.xlabel('Month')
plt.ylabel('Total Revenue')
plt.xticks(rotation=45)
plt.grid(True, linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()
```

```python
# Distribution of total spending
plt.figure(figsize=(10,6))

sns. histplot(
    df['Total Spent'],
    bins=15,          # fewer bins = wider bars
    kde=False,        # remove the smooth KDE line
    color='darkred',  # deep red color
    edgecolor='black' # black border for each bar
)

plt.title('Distribution of Total Spending', fontsize=16, fontweight='bold')
plt.xlabel('Total Spent', fontsize=12)
plt.ylabel('Frequency', fontsize=12)
plt.tight_layout()
plt.show()

```

```python
#Online vs In-store (Offline) Shopping Revenue Graph

#Group data
sales_by_location = df.groupby('Location')['Total Spent'].sum().reset_index()

#Define color palette based on actual categories
palette_colors = {
    'Online': 'red',
    'In-store': 'yellow',
    'Offline': 'yellow'  # just in case 'Offline' appears in another dataset
}

#Plot
plt.figure(figsize=(8,6))
sns.barplot(
    x='Location',
    y='Total Spent',
    hue='Location',
    data=sales_by_location,
    palette=palette_colors,
    legend=False,
    edgecolor='black'
)

#Chart
plt.title('Online vs In-store Shopping Revenue', fontsize=16, fontweight='bold')
plt.xlabel('Shopping Type', fontsize=12)
plt.ylabel('Total Revenue', fontsize=12)
plt.tight_layout()
plt.show()

```
